```python
import streamlit as st
import requests
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

FLASK_URL = "http://localhost:5000"

st.set_page_config(page_title="Smart Salary Predictor", page_icon="💼", layout="wide")

st.title("Smart Salary Predictor")
st.caption("Powered by Linear Regression + Claude AI suggestions")

# ── Sidebar inputs ──────────────────────────────────────────────
with st.sidebar:
    st.header("Your Profile")

    job_title = st.selectbox("Job title", [
        "Data Scientist", "Data Engineer", "ML Engineer",
        "Data Analyst", "Research Scientist", "Analytics Engineer",
        "Data Architect", "BI Developer", "Other"
    ])

    experience_level = st.selectbox("Experience level", [
        "Entry", "Mid", "Senior", "Executive"
    ])

    employment_type = st.selectbox("Employment type", [
        "Full-time", "Part-time", "Contract", "Freelance"
    ])

    company_size = st.selectbox("Company size", ["Small", "Medium", "Large"])

    remote_ratio = st.select_slider(
        "Remote ratio (%)", options=[0, 50, 100], value=50
    )

    residence = st.text_input("Your country code (e.g. US, IN, GB)", value="US").upper()
    company_location = st.text_input("Company country code", value="US").upper()
    work_year = st.selectbox("Work year", [2020, 2021, 2022, 2023])

    predict_btn = st.button("Predict Salary", type="primary", use_container_width=True)

# ── Main panel ──────────────────────────────────────────────────
col1, col2 = st.columns([1, 1])

if predict_btn:
    payload = {
        "job_title": job_title,
        "experience_level": experience_level,
        "employment_type": employment_type,
        "company_size": company_size,
        "remote_ratio": remote_ratio,
        "residence": residence,
        "company_location": company_location,
        "work_year": work_year,
    }

    with st.spinner("Predicting..."):
        try:
            resp = requests.post(f"{FLASK_URL}/predict", json=payload, timeout=10)
            predicted = resp.json()["predicted_salary"]
        except Exception as e:
            st.error(f"Flask server not running. Start it with: python app.py\n{e}")
            st.stop()

    with col1:
        st.subheader("Predicted Salary")
        st.metric("Annual (USD)", f"${predicted:,.0f}")

        # Salary gauge chart
        avg_salary = 112000  # approximate dataset average
        fig = go.Figure(go.Indicator(
            mode="gauge+number+delta",
            value=predicted,
            delta={"reference": avg_salary, "prefix": "$", "valueformat": ",.0f"},
            title={"text": "vs. market average"},
            gauge={
                "axis": {"range": [0, 300000]},
                "bar": {"color": "#534AB7"},
                "steps": [
                    {"range": [0, 80000],   "color": "#F0997B"},
                    {"range": [80000, 140000], "color": "#9FE1CB"},
                    {"range": [140000, 300000], "color": "#5DCAA5"},
                ],
                "threshold": {
                    "line": {"color": "#0F6E56", "width": 3},
                    "thickness": 0.75,
                    "value": avg_salary
                }
            }
        ))
        fig.update_layout(height=300, margin=dict(t=40, b=0, l=20, r=20))
        st.plotly_chart(fig, use_container_width=True)

    with col2:
        st.subheader("Salary by Experience")
        # Show how salary scales with experience for context
        exp_data = pd.DataFrame({
            "Experience": ["Entry", "Mid", "Senior", "Executive"],
            "Avg Salary": [61000, 94000, 131000, 168000]
        })
        fig2 = px.bar(
            exp_data, x="Experience", y="Avg Salary",
            color="Avg Salary", color_continuous_scale="Teal",
            text_auto="$,.0f"
        )
        fig2.update_layout(
            height=300, showlegend=False,
            margin=dict(t=20, b=0, l=0, r=0),
            coloraxis_showscale=False
        )
        st.plotly_chart(fig2, use_container_width=True)

    # ── AI Suggestions ──────────────────────────────────────────
    st.divider()
    st.subheader("AI Career Suggestions")
    st.caption("Powered by Claude — personalized advice to boost your salary")

    with st.spinner("Getting AI suggestions..."):
        try:
            suggest_payload = {**payload, "predicted_salary": predicted}
            resp2 = requests.post(f"{FLASK_URL}/suggest", json=suggest_payload, timeout=30)
            suggestions = resp2.json()["suggestions"]
            st.info(suggestions)
        except Exception as e:
            st.warning(f"Could not get AI suggestions: {e}")

else:
    with col1:
        st.info("Fill in your profile in the sidebar and click **Predict Salary**.")
    with col2:
        st.markdown("**What this app does:**")
        st.markdown("""
- Predicts your salary using a Linear Regression model trained on 607 real data science job salaries
- Shows how you compare to market average
- Gives personalized AI suggestions (powered by Claude) to grow your salary
        """)
```
